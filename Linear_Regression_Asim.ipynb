```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
#import math
```

```python
#Reading Data
data = pd.read_excel('swedish.xls')
print(data.shape)
data.head()
```

```python
# Collecting X & Y
X = data['X'].values
Y = data['Y'].values
```

```python
# Calculate Mean of X & Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

# Calculate Total No. of values
n = len(X)
print(n, mean_x, mean_y)
```

```python
# Using the Formula to Calculate the Slope and C
asim = 0
asin = 0
for i in range(n):
    asim += (X[i] - mean_x) * (Y[i] - mean_y)
    asin += (X[i] - mean_x) ** 2
b1 = asim/asin  #Slope or m
b0 = mean_y - (b1 * mean_x)  #C

#Print Coeff.
print(b1, b0)
    
```

```python
# Plotting Values & Regresion Line

max_x = np.max(X) + 100
min_x = np.min(X) - 100

# Calculating line values X and Y
x = np.linspace(min_x, max_x, 400)
y = b0 +b1 * x

# Plotting Line
plt.plot(x, y, label = 'Regression Line')
# Plotting Scatter Points
plt.scatter(X, Y, label = 'Scatter Plot')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0
for i in range(n):
    y_pred = b0 + b1 * x[i]
    ss_t += (Y[i] - mean_y) ** 2
    ss_r += (Y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)
print(r2)
```
